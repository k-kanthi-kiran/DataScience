$Value-Counts$

```python
# How many asia count is available out of 25k entries
# How many africa count is available out of 25k entries. 
visadf['continent'].value_counts() # returns a series 
visadf['continent'].value_counts(normalize=True)
continent_Values=dict(visadf['continent'].value_counts()) # converting to dictionary 
keys=list(continent_Values.keys())
values=list(continent_Values.values())
continent_df=pd.DataFrame(zip(keys,values)) # creating a data frame
continent_df=pd.DataFrame(zip(keys,values),columns=['continent','count']) #creating a data frame with 2 columns only 
continent_df.to_csv('continent_csv',index=False)
```

```python
# Value error: If using all scalar values, you must pass an index
continent_values=dict(visadf['continent'].value_counts())
pd.DataFrame(continent_values,index=['Count']) # if you dont pass the index over here you should get the above error. 
# In the dictionary if values in the form of list: then no need to pass the index. 
# In the dictionary if values are as single(scalar) then provide index. 
#continent_df is a frequency table.
```

$Bar-Chart$

```python
#Bar chart requires one categorical column  and one numerical column 
# continent df output consider as frequency table. 
# IN that we have one cat column : Continent( X -axis)
# count is a numerical column : (Y-Axis)
continent_df
plt.figure(figsize=(9,5))
plt.bar('continent','count',data=continent_df)
plt.xlabel('Continents')
plt.ylabel('Count')
plt.title("Bar Chart")
plt.savefig("Continent_bar_chart.jpg")
plt.show()
```
