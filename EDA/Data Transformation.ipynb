**Reciprocal Transformation**

Receiprocal transformation fails when data has zero values. 

```python
recipro_data= np.reciprocal(exp_data)
plt.figure(figsize=(10,5))
plt.subplot(1,2,2).hist(recipro_data,bins=40,label='Reciprocal Transformation')
plt.legend()
plt.show()
```

**Square root Transformation**

**Power transformation**
- It is related to sklearn package
- Package name:sklearn.preprocessing
- Method name: Power Transformer
- Inside Box-Cox,yeo-jhonson

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
pt_data = pt.fit([[exp_data]]).transform([[exp_data]])
pt_data
```
