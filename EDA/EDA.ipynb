```python
# Method 1
# Creating a dataframe using list
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
pd.DataFrame(zip(Names,Age))
```

```python
#With Columns
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
pd.DataFrame(zip(Names,Age),columns=['Name','Age'])
```

```python
#With Index
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
pd.DataFrame(zip(Names,Age),columns=['Name','Age'],index=['A','B',100])
```

```python
#Create a new column and Adding to the frame
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),columns=['Name','Age'],index=['A','B',100])
df1['City']=['Bangalore','Hyderabad','Chennai']
df1
```

```python
#Updating the data in the entire column 
df1['Name']=['Rambo','Arnold','Alex']
df1
```

```python
#Adding a row to dataframe using loc
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),columns=['Name','Age'],index=['A','B',100])
df1.index
len(df1.index)
```

```python
#Changing the index from 'A','B',100 to 1,2,3 and Adding a row to dataframe
df1=pd.DataFrame(zip(Names,Age),columns=['Name','Age'],index=[0,1,2])
df1.loc[3]=['Bunny',40]
df1
df1.loc['A']=['Sunny',50]
df1
```

```python
#Dropping the column using drop method which can be used for dropping column or row. 
# In this AXIS=1 references column and AXIS=0 references rows
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),columns=['Name','Age'],index=[1,2,3])
df1['City']=['Bangalore','Hyderabad','Chennai']
df1.loc[3]=['Bunny',40,'Delhi']
df1.loc[4]=['Sunny',35,'Bihar']
df1.drop('City',axis=1) # Dropping city column , The data frame is not modified and on printing we get back the original
df1.drop('City',axis=1,inplace=True)
df1
```

```python
df1.drop(1,inplace=True)
```
