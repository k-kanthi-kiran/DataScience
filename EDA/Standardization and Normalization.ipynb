**Standardization and Normalization**
 - Standardization and Normalization are two approaches to convert data values to same range/scale.It is also scaling of data. 
 - In the data we have many columns are there in which each column has different units as well as differnt values. 
 - For ex: We have age and salary. Age is minimal number basically a two digit number and salary is a 5 digit number.
 - When you multiply two digit number(age) with the 5 digit number(salary) it involves some complexity. 
 - Imagine multiplying both are single digit number, this involves less complexity
 - Scaling converts all the data into the same scale
 - In which one technique is standardization which is also called Z-Scale or Z-Score.
 - It ranges -3 to 3
 - The mean=0 and standard deviation=1
 
 
 - **Normalization**
 - Normalization converts data into 0 to 1 range.
 - minimum =0 and maximum value to 1
 - It is mainly used in deep learning project for image scaling
 - Generally images are color images and pixel value ranges from 0 to 255
 - we Normalize the value from 0 to 1
 - min max scaler
 - The value might change but the information will not change. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filelocation="C:\\Users\\ADMIN\\Desktop\\Python\\Files\\Visadataset.csv"
visadf = pd.read_csv(filelocation)
visadf
```

```python
** Z-score**
- First we read a specific column 
- Calculate the mean of the column 
- Calculate the standard deviation of the column 
- Numerator = column - mean value
- Numerator/standard deviation. 
```

```python
mean=visadf['prevailing_wage'].mean()
deviation = visadf['prevailing_wage'].std()
numerator = visadf['prevailing_wage']-mean
visadf['prevailing_wage_zscore']=numerator/deviation
visadf[['prevailing_wage_zscore']],visadf[['prevailing_wage']]
```

```python
max_original=visadf['prevailing_wage'].max()
max_zoriginal=visadf['prevailing_wage_zscore'].max()
max_original,max_zoriginal
```

```python
#Identifying the row id  with maximum prevailing wage
print(visadf['prevailing_wage'].idxmax())
print(visadf['prevailing_wage_zscore'].idxmax())
```

```python
#Finding minimum
min_original=visadf['prevailing_wage'].min()
min_zoriginal=visadf['prevailing_wage_zscore'].min()
min_original,min_zoriginal
print(visadf['prevailing_wage'].idxmin())
print(visadf['prevailing_wage_zscore'].idxmin())
```

**Z score is also used for identifying the Influential Outliers**

- Generally outlier means very very huge observations or very very small observation
- If we found an observation as outlier before scaling
- The same observation again found as outlier after scaling
- Then that observation called influential outlier
- Some observation before scaling consider as outlier
- But after scaling  it does not fall in outlier criteria
- At that time we might not consider that observation as outlier
- Z score is used to identify influential outliers

```python
# Below you can see that both the outliers are the same. 
plt.subplot(1,2,1)
plt.boxplot(visadf['prevailing_wage']) 
plt.subplot(1,2,2)
plt.boxplot(visadf['prevailing_wage_zscore'])
```

**Min max scalar**
- Read the column 
- Calculate the min value
- Calculate the max value
- Numerator = column - minimum value
- denomiator = maximum - minimum 
- output = Numerator/denominator



```python
filelocation="C:\\Users\\ADMIN\\Desktop\\Python\\Files\\Visadataset.csv"
visadf = pd.read_csv(filelocation)

min_original=visadf['prevailing_wage'].min()
max_original = visadf['prevailing_wage'].max()
 

Numerator  = visadf['prevailing_wage'] - min_original
denomiator =  max_original - min_original
output = Numerator/denomiator
visadf['prevaling_wage_norm']=output
```

```python
print(visadf['prevailing_wage'].idxmin())
print(visadf['prevaling_wage_norm'].idxmax())
```
