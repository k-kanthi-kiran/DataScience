```python
# If we want to separate numerical and categorical colums 
#- Convert the data frame to dictionary
visadict=dict(visadf.dtypes)

for i in visadict:
    if visadict[i] == 'object':
        print(i,visadict[i])
    
categorical_columns=[i for i in visadict if visadict[i] == 'object']
numerical_columns=[i for i in visadict if visadict[i] != 'object']
print("Categorical Columns are :",categorical_columns)
print("Numerical Columns are :",numerical_columns)
```

```python
# Other way of fetching categorical and Numerical columns : returns dataframe
print("Categorical Columns are :",visadf.select_dtypes(include='object').columns)
print("Numerical Columns are :",visadf.select_dtypes(exclude='object').columns)
```

```python
#when ever we open the notebook read the packages
filelocation="C:\\Users\\ADMIN\\Desktop\\Python\\Files\\Visadataset.csv"
visadf = pd.read_csv(filelocation)
visadf
#visadf.iloc[<rows>,<columns>]
#visadf.iloc[<start:end>,<start:end>]
#visadf.iloc[<rows index>,<column index>]
#visadf.iloc[5:10,2:6] # want to select rows has index 5 to 9 and Columns 2 to 5
```

```python
visadf.iloc[:,2:9] # All the rows from 2nd column to 8th Column
```

```python
#Specific rows and columns using iloc
rows=[100,200,300]
cols=[2,7]
visadf.iloc[rows,cols]
```

```python
#Specific rows and columns with column name using loc
#Difference between iloc and loc is that iloc takes only numerical entry we cannot provide column names directly whereas 
# loc does not take the numerical index it takes only column names. 
rows=[100,200,300]
cols=['education_of_employee','region_of_employment']
visadf.loc[rows,cols]
```
